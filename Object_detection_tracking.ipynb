```python
#from zmq import device

device= 'cuda'

model=torch.hub.load('ultralytics/yolov5', 'yolov5s')
model.to(device)
```

```python
#Test to label

prediction = model('./yolov5/data/images/bus.jpg')
print(prediction.pandas().xyxy[0]['ymin'])
```

```python
def get_coordenates(prediction):
    xmin=prediction.pandas().xyxy[0]['xmin']
    ymin=prediction.pandas().xyxy[0]['ymin']
    xmax=prediction.pandas().xyxy[0]['xmax']
    ymax=prediction.pandas().xyxy[0]['ymax']
    confidence=prediction.pandas().xyxy[0]['confidence']
    class_number=prediction.pandas().xyxy[0]['class']
    class_name=prediction.pandas().xyxy[0]['name']
    return xmin,ymin, xmax, ymax, confidence, class_number, class_name
```

Open the video to be used

```python
from itertools import count
from re import X
from statistics import mode
from termios import FF0
import cv2 

video=cv2.VideoCapture('./video/Road traffic video for object recognition.mp4')

while True:
    r, frame = video.read()
    count=0
    if r is True:

        prediction=model(frame)
        xmin,ymin, xmax, ymax, confidence, class_number, class_name=get_coordenates(prediction)    
        for class_numbe_id in range(class_number.shape[0]):
            
            if class_name[class_numbe_id]=='car' or class_name[class_numbe_id]=='truck':
                cv2.putText(frame, 'Number of vehicles:{}'.format(int(count)), org=(10,120), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=2, color=(0,255,0), thickness=1)

                #Identify the objects
                cv2.putText(frame, 'Class name:{}, Score:{}'.format(class_name[class_numbe_id], round(float(confidence[class_numbe_id]), 3)), org=(int(xmin[class_numbe_id]),int(ymin[class_numbe_id])-15), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1, color=(225,0,0), thickness=1)
                cv2.rectangle(frame,pt1=(int(xmin[class_numbe_id]),int(ymin[class_numbe_id])), pt2=(int(xmax[class_numbe_id]),int(ymax[class_numbe_id])), color=(255,0,0))
                count+=1



        if cv2.waitKey(25) & 0xFF==ord('q'):
            break

    cv2.imshow('Original video',frame)

video.release()
cv2.destroyAllWindows()
```
